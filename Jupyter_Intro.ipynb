### Functions

Define a simple python function and print the result

````python
def sum_of_a(a):
    """
    Finds sum of all natural numbers till input 'a'
    """
    total = 0
    for i in range(a+1):
        total+=i
    return total
````

````python
result = sum_of_a(10)
print(result)
print('The result is:',result)
print('The result is: '+str(result))
````

### Line and Cell magic
1. Line magics: these are commands prepended by one % character and whose arguments only extend to the end of the current line.
2. Cell magics: these use two percent characters as a marker (%%), and they receive as argument both the current line where they are declared and the whole body of the cell. Note that cell magics can only be used as the first line in a cell, and as a general principle they can't be 'stacked' (i.e. you can only use one cell magic per cell). A few of them, because of how they operate, can be stacked, but that is something you will discover on a case by case basis.


````python
# List all available magics, and it will show both line and cell magics currently defined
%lsmagic 
````

## Plotting

Visualization tools explained using sample dataset

````python
%matplotlib inline
import seaborn as sns
data = sns.load_dataset("tips")
````

````python
data.head() #displays first 5 rows of the dataset
````

````python
data.describe() #gives statisctics about the data attributes
````

Plotting attribute **subplots** as a single figure. See comments for detailed description 

````python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15,3] # setting display size
attr = ['total_bill', 'tip', 'size']

fig, ax = plt.subplots(nrows=1, ncols=len(attr)) # variable column size based on attributes

i = 0
for col in ax:
    col.title.set_text(attr[i]) # subplot title
    col.plot(data[attr[i]])
    i += 1
````

Scatter plots using seaborn package

````python
plt.rcParams['figure.figsize'] = [8,5]
sns.scatterplot(x="total_bill", y="tip", data=data);
````

````python
sns.scatterplot(x="total_bill", y="tip", hue="size", style="smoker", data=data);
plt.savefig("test.svg")
````

````python
sns.regplot(x="total_bill", y="tip", data=data);
````
